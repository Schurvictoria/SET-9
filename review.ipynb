```python
def parse_data(raw_text):
    lines = raw_text.strip().split('\n')
    data = []
    current_array_type = "Initial"
    current_size = None

    for line in lines:
        line = line.strip()
        if "array size" in line:
            match = re.match(r'(.+?) array size (\d+)', line)
            if match:
                current_array_type = match.group(1)
                current_size = int(match.group(2))
        else:
            match = re.match(r'(.+?)\s+Time:\s+(\d+)\s+ms\s+Char comparisons:\s+(\d+)', line)
            if match:
                algorithm = match.group(1).strip()
                time_ms = int(match.group(2))
                char_comparisons = int(match.group(3))
                data.append({
                    'Algorithm': algorithm,
                    'ArrayType': current_array_type,
                    'ArraySize': current_size,
                    'Time_ms': time_ms,
                    'CharComparisons': char_comparisons
                })
    return pd.DataFrame(data)
```

```python
df = parse_data(data)

summary = df.groupby(['Algorithm', 'ArrayType']).agg({
    'Time_ms': 'mean',
    'CharComparisons': 'mean',
    'ArraySize': 'mean'
}).reset_index()
```

```python
plt.figure(figsize=(14,8))
sns.lineplot(data=df, x='ArraySize', y='Time_ms', hue='Algorithm', style='ArrayType', markers=True)
plt.title('Время работы алгоритмов сортировки')
plt.xlabel('Размер массива')
plt.ylabel('Время (мс)')
plt.legend(title='Алгоритм / Тип массива', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

На этом графике показано, как меняется время работы различных алгоритмов сортировки в зависимости от размера массива. Заметим, что:

Ternary QuickSort значительно медленнее остальных алгоритмов, особенно на обратносортированных массивах — время растёт почти экспоненциально.

QuickSort, MergeSort, использующий длину наибольшего префикса, показывают стабильное и низкое время работы, практически не зависящее от типа массива (линию Quick sort не видно, однако предполагаю, что она перекрыта красной, оранжевой и зеленой линией. Поэтому скорее всего по времени будет примерно такая же)

MSD Radix Sort с переключением (для понимания - cutoff) демонстрирует минимальное время работы для всех массивов — линия идёт у самого низа графика.

MSD Radix Sort без переключения (для понимания - pure) также показывает очень быстрое выполнение, но чуть медленнее оптимизированной версии, особенно на больших массивах. Это связано с тем, что без переключения на быструю сортировку для маленьких подмассивов неоптимизированная версия без переключения тратит больше времени на глубокую рекурсию.
